```
# Check if this file is present in /FileStore/tables/

# DO NOT CHANGE THE CONTENT OF THIS CELL
def file_exists(path):
    try:
        dbutils.fs.ls(path)
        return True
    except Exception as e:
        if 'java.io.FileNotFoundException' in str(e):
            return False
        else:
            raise

try:
    imp.find_module('dbutils')     
    assert file_exists("/FileStore/tables/" + zip_file) == True, "ERROR - You didn't upload the zip_file, please upload into /FileStore/tables/ "
except ImportError:
    print("Are you running this on Databricks?")
```

# Copy from dbfs to file:

```
dbutils.fs.cp(fileloc, "file:/tmp/zip_files/"+zip_file)
```

```
dbutils.fs.ls("file:/tmp/zip_files/")
```

# Create environ variable

```
os.environ['fileroot'] = zip_file
os.environ['fileroot2'] = file_name

print(os.environ['fileroot'])
print(os.environ['fileroot2'])
```

```
%sh

ls /tmp/zip_files/
```

# Unzip the .zip file into /extracts/ folder

```
%sh

unzip -d /tmp/zip_files/extracts/ /tmp/zip_files/$fileroot
```

```
%sh
ls /tmp/zip_files/extracts/$fileroot2
```

# Move it back to dbfs

```
dbutils.fs.cp("file:/tmp/zip_files/extracts/"+file_name, filestore+"/"+file_name, True)
```

```
dbutils.fs.ls(filestore+"/"+file_name)
```
