--> 14. Now define a new RDD containing just the IP addresses from each line in the log file. (The IP address is the first
"word" in each line).

```
logfile = "/databricks-datasets/sample_logs/part-00000"               # logfile variable creation
RDD_log_file = sc.textFile(logfile)                                   # RDD initialization

RDD_lf_array = RDD_log_file.map(lambda line: line.split(' '))         # RDD_array definition by splitting the data

IP_RDD = RDD_lf_array.map(lambda item: item[0])                       # Extraction of 1st element from each line into the new IP RDD
IP_RDD.collect()
```
