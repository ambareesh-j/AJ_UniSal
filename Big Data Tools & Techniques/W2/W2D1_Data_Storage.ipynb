Access files using DBFS (dbutils.fs) (Default dbfs:/)

1. Local files

```
print("Local files:")
dbutils.fs.ls("file:/tmp")
```

2 DBFS files

```
print("DBFS files:")
dbutils.fs.ls("dbfs:/")
```

```
print("DBFS files:")
dbutils.fs.ls("dbfs:/FileStore/tables")
```

```
# By default it assumes dbfs: parameter

print("DBFS files:")
dbutils.fs.ls("/FileStore/tables")
```
