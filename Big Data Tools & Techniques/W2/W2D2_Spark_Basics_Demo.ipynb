## Create an RDD using the textFile() method & parallelize() method

```
mydata = sc.textFile("dbfs:/FileStore/tables/Dummy.txt")
mydata.count()
```

## RDD.take() method

```
# mydata = sc.textFile("dbfs:/FileStore/tables/Dummy.txt")    # -> Already given above
print(mydata.count())                                         # -> Explicit print
for line in mydata.take(2):                                   # take() function is simply like head() function
  print(line)
```

## Transformations - map() & filter() methods - using Anonymous functions

```
# mydata = sc.textFile("dbfs:/FileStore/tables/Dummy.txt")    # -> Already given above
mydata_upper = mydata.map(lambda line: line.upper())
print(mydata_upper)

mydata_filter = mydata.filter(lambda line: line.startswith ("U"))
mydata_filter.count()
```

## Another example - take()

```
 def to_Upper(s):
    return s.upper()

mydata = sc.textFile("dbfs:/FileStore/tables/Dummy.txt")
mydata.map(to_Upper).take(2)
```

```
# Using lambda function directly to make it an anonymous function

mydata = sc.textFile("dbfs:/FileStore/tables/Dummy.txt")
mydata.map(lambda line: line.upper()).take(2)
```
