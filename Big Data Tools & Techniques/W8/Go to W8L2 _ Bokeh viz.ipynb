#### Bokeh: Data Visualization

Unlike matplotlib or seaborn, Bokeh renders in HTML and Javascript.
These are interactive plots (not static like other 2 libs)

```
%pip install bokeh
```

```
#https://docs.databricks.com/_static/notebooks/bokeh-demo.html
from bokeh.plotting import figure
from bokeh.embed import components, file_html
from bokeh.resources import CDN

# prepare some data
x = [1,2,3,4,5]
y = [6,7,2,4,5]

# create a new plot with a title and axes labels
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.line(x,y,legend_label="Temp.", line_width=2)

#create an html document that embeds the Bokeh plot
html = file_html(p, CDN, "my plot1")

# display this html
displayHTML(html)
```

```
# bokeh basics
from bokeh.plotting import figure
from bokeh.io import show, output_notebook

# Create a blank figure with labels
p = figure(plot_width = 600, plot_height = 600,
           title = 'Example',
           x_axis_label='x', y_axis_label='y')

# Example data
squares_x = [1,3,4,5,8]
squares_y = [8,7,3,1,10]
circles_x = [9,12,4,3,15]
circles_y = [8,4,11,6,10]

# Add squares 
p.square(squares_x, squares_y, size = 12, color = 'navy', alpha = 0.6)

# Add circle 
p.circle(circles_x, circles_y, size =12, color = 'red')

# Create an html document that embeds the Bokeh plot
html = file_html(p, CDN, "my plot1")

# Display this html
displayHTML(html)
```
