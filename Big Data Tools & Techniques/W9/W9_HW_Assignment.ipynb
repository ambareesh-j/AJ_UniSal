##### --> 4. Produce a csvFilesRDD, which only contains the FileInfo objects whose filename ends with .csv. NB: The items in

##### --> 5. The format of the file names is MM-DD-YYYY.csv where MM stands for month, DD day and YYYY year. Sorting based on
this type of timestamp directly would not necessarily give us the desired results, so we need to do two things:
(a) Reorder the timestamp so that we're guaranteed the natural order is the order we want, i.e. we want to take
MM-DD-YYYY.csv to YYYYMMDD.
(b) To stop string based sorting, you need to cast the new timestamp to integer. The following function will do this:

```
import re

# Assumes input MM-DD-YYYY.csv
def convert_name(filename):
  m = re.match("^([0-9]{2})-([0-9]{2})-([0-9]{4}).csv$", filename)
  return int(m.group(3)+m.group(1)+m.group(2))
```

```
csvPairRDD = csvFilesRDD.map(lambda line: (convert_name(line[1]) , line[0]))

csvPairRDD.take(5)
```

##### --> 6. Using sortByKey nd the most recent pair (i.e. the pair which has the most recent value as its key).

```
csvPairRDD.sortByKey(ascending=False).take(5)
```

##### --> 7. Assign the path of the most recent pair to the variable chosenfile.

```
chosenfile = csvPairRDD.sortByKey(ascending=False).map(lambda line: line[1]).take(1)

print(chosenfile)
```

##### --> 8. Load chosenfile into a Spark dataframe. You should acount for the fact that the header forms the rst line of the data.

```
# File location and type
file_location = chosenfile
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df_chosen = spark.read.format(file_type) \
               .option("inferSchema", infer_schema) \
               .option("header", first_row_is_header) \
               .option("sep", delimiter) \
               .load(file_location)

# Dataframe df_cord displayed
df_chosen.printSchema()
```
